```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
dogs_image = plt.imread("/content/2_Lovely_Dog_Puppy.jpg")

numpay_image=Image.fromarray(dogs_image)
plt.imshow(dogs_image)
```

```python
a1,a2,a3, a4, a5, a6 = np.split(dogs_image, 6)
plt.imshow(a1)
plt.show()
plt.imshow(a2)
plt.show()
plt.imshow(a3)
plt.show()
plt.imshow(a4)
plt.show()
plt.imshow(a5)
plt.show()
plt.imshow(a6)
plt.show()
```

```python
numpay_image=Image.fromarray(np.concatenate((a2,a4,a6)))
plt.imshow(np.concatenate((a2,a4,a6)))
```
